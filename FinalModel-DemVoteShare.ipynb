```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import Lasso
```

```python
df1 = pd.read_csv("finaldataset.csv")

# print first few rows -- could also use df.head()
print (df1.iloc[:3])
# and get summary stats on variables
print (df1.describe())
```

```python
df2 = df1.dropna()
# and get summary stats on variables to check to see if any variables had missingness
print (df2.describe())
```

```python
# LIN REGRESSION
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

dataset = df2

X = dataset[['incumbency','consec_terms','home_state','black','hispanic','cuban','nonwhite','pca1','evangelical','educated','pca2','polling_variable','third_party','past_elections']]

y = dataset['vote_dem']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.99, test_size=0.01, random_state=7)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)

from sklearn import metrics
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))

r_sq = regressor.score(X_train, y_train)
print("R^2: " + str(r_sq))
```

```python
coeff_df = pd.DataFrame(regressor.coef_, X.columns, columns=['Coefficient'])
coeff_df
```
